```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
print(len(df_train), len(df_test))
df_train
```

```python
# 널값 처리, 버릴거 버리기 
drop_cols = ["Origin_Airport", "Destination_Airport","Carrier_Code(IATA)", "ID", "Cancelled", "Diverted"]
df_train.drop(drop_cols, axis=1,inplace=True)
df_test.drop(drop_cols, axis=1,inplace=True)
df_train.info()
```

```python
# Delay Null값 버리기..
df_train_drop = df_train.dropna(subset=["Delay"], axis=0)

df_train_drop.info()
```

```python
df_train_drop.isnull().sum()
```

```python
# 1. 각 Origin_Airport_ID별로 가장 많이 등장하는 Origin_State 찾기
mode_states = df_train_drop.groupby("Origin_Airport_ID")["Origin_State"].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# 2. Origin_State가 NaN인 데이터에 대해 Origin_Airport_ID를 기준으로 값 채우기
df_train_drop["Origin_State"] = df_train_drop.apply(
    lambda row: mode_states[row["Origin_Airport_ID"]] if pd.isnull(row["Origin_State"]) else row["Origin_State"],
    axis=1
)

# 1. 각 Origin_Airport_ID별로 가장 많이 등장하는 Origin_State 찾기
mode_states = df_test.groupby("Origin_Airport_ID")["Origin_State"].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# 2. Origin_State가 NaN인 데이터에 대해 Origin_Airport_ID를 기준으로 값 채우기
df_test["Origin_State"] = df_test.apply(
    lambda row: mode_states[row["Origin_Airport_ID"]] if pd.isnull(row["Origin_State"]) else row["Origin_State"],
    axis=1
)

df_train_drop.isnull().sum()
```

```python
mode_states = df_train_drop.groupby("Destination_Airport_ID")["Destination_State"].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
mode_states
```

```python
df_train_drop["Destination_State"] = df_train_drop.apply(
    lambda row: mode_states[row["Destination_Airport_ID"]] if pd.isnull(row["Destination_State"]) else row["Destination_State"],
    axis=1
)

# test 데이터에 대해 Destination_State가 NaN인 경우 최빈값으로 채우기
df_test["Destination_State"] = df_test.apply(
    lambda row: mode_states.get(row["Destination_Airport_ID"], None) if pd.isnull(row["Destination_State"]) else row["Destination_State"],
    axis=1
)

df_train_drop.isnull().sum()
```

```python
df_train_dropped = df_train_drop.dropna(axis=1)
df_train_dropped.info()
```

```python
df_test.info()
```

```python
df_test = df_test.apply(lambda x : x.fillna(x.mode()[0]))
df_test.info()
```

```python
# ID는 필요없음
# Month, Day_of_Month는 일자
# Estimated_Departure_Time, Estimated_Arrival_Time : 출발 시간 HH:MM 형식으로 변경할 필요 있음
# Cancelled, Diverted : 바이너리
# Airport = Airport_ID, Origin_State-> 고유값 확인 필요
# Destination도 동일
# Distance 중요할 것이라생각함
# Airline 항공사 고유값 확인 필요
# Carrier_Code=Carrier_ID : 항공사 고유 코드 -> 고유값 확인 필요
# Tail_Number : 항공기 고유 번호 
# null_cols = df_train.columns[df_train.isnull().sum() > 0].to_list()
df_train_dropped.apply(lambda x:x.unique())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

mapped = {"Not_Delayed" : 0, "Delayed" : 1}
df_train_dropped["Delay"] = df_train_dropped["Delay"].map(mapped)
```

```python
lbl_cols = ["Origin_State", "Destination_Airport_ID", "Destination_State", "Tail_Number"]
def unknown_encoding(value, encoder, unknown=-1):
    if value in encoder.classes_:
        return encoder.transform([value])[0]
    else:
        return unknown
        
for i in lbl_cols:
    lbl_encoder = LabelEncoder()
    df_train_dropped[i] = lbl_encoder.fit_transform(df_train_dropped[i])
    df_test[i] = df_test[i].apply(lambda x: unknown_encoding(x, lbl_encoder, -1))
df_train_dropped.head()
```
