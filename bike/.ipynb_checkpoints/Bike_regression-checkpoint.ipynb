```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler, MinMaxScaler
```

```python
df_test_drop = df_test.drop(["datetime", "minute", "second"], axis=1)
```

```python
# df_train_drop = df_train_drop.drop(["dayofweek"], axis=1)
# df_test_drop = df_test_drop.drop(["dayofweek"], axis=1)

mul_cols = ["season", "holiday", "workingday", "weather", "year"]
encoder = OneHotEncoder(sparse_output=False)
train_encoded = encoder.fit_transform(df_train_drop[mul_cols])
test_encoded = encoder.transform(df_test_drop[mul_cols])

encoded_cols = encoder.get_feature_names_out(mul_cols)
print(encoded_cols)
train_encoded = pd.DataFrame(train_encoded, columns=encoded_cols)
test_encoded = pd.DataFrame(test_encoded, columns=encoded_cols)
train_encoded
```

```python
df_train_encoded = df_train_drop.drop(mul_cols, axis=1)
df_train_encoded = pd.concat([df_train_encoded, train_encoded], axis=1)

df_test_encoded = df_test_drop.drop(mul_cols, axis=1)
df_test_encoded = pd.concat([df_test_encoded, test_encoded], axis=1)
```

```python
df_train_encoded.info()
```

```python
scaler = MinMaxScaler()
mul_cols = ["temp", "atemp", "windspeed", "humidity"]
df_train_encoded[mul_cols] = scaler.fit_transform(df_train_encoded[mul_cols])
df_test_encoded[mul_cols] = scaler.transform(df_test_encoded[mul_cols])
```

```python
df_train_encoded
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import accuracy_score as acc
from sklearn.model_selection import train_test_split
```

```python
y = df_train_encoded[["count"]]
x = df_train_encoded.drop(["count"], axis=1)

x_test = df_test_encoded
print(x.head())
print(y.head())
```

```python
# 학습 후 결과가 음수가 나왔음, 이는 0과 양수 내에서 예측할 수 있도록 해야함 -> 로그변환을 사용ㅇ
y = np.log1p(y)
print(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
print(len(X_train))
print(len(X_val))
```

```python
gb = GradientBoostingRegressor(
    n_estimators = 200,
    max_depth = 4,
    learning_rate = 0.05,
    random_state = 42
)
gb.fit(X_train, y_train)
```

```python
y_predict = gb.predict(X_val)
y_predict = np.expm1(y_predict)
print(y_predict)
```

```python
train_score = gb.score(X_train, y_train)
val_score = gb.score(X_val, y_val)
print(train_score, val_score)
```

```python
# test 제출
x_test.head()
```

```python
test_predict = gb.predict(x_test)
test_predict = np.expm1(test_predict)
```
