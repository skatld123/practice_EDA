```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
data= fetch_california_housing(as_frame = True)
data
```

```python
#데이터 생성
feature= pd.DataFrame(data.data)
feature.columns = data.feature_names

target = pd.DataFrame(data.target)
target.columns=data.target_names
```

```python
df = pd.concat([feature, target],axis= 1)
df.head()
```

```python
df.hist(figsize=(12,10))
```

```python
df.isnull().sum()
```

- AveRooms, AveBedrms, Population, AveOccup 이 Skew함으로, log변환이 필요

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
print(corr["MedHouseVal"].abs().sort_values(ascending=False))
```

```python
sns.scatterplot(data=df, x="Longitude", y="Latitude", size="MedHouseVal", hue="MedHouseVal")
```

```python
log_list = ["AveRooms", "AveBedrms", "Population", "AveOccup"]
for i in log_list:
    df[i] = np.log1p(df[i]) # 역변환은 expm1
df[log_list].hist()
```

```python
from sklearn.cluster import KMeans

# K-means 클러스터링을 통해 좌표 기반 클러스터 생성
kmeans = KMeans(n_clusters=5, random_state=0)
df['location_cluster'] = kmeans.fit_predict(df[['Latitude', 'Longitude']])
```

```python
sns.scatterplot(data=df, x="Longitude", y="Latitude", size="MedHouseVal", hue="location_cluster")
```

```python
abs(df.corr()["MedHouseVal"])>0.1
```
