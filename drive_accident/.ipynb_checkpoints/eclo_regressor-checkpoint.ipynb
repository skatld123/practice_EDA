```python
df_train["사고일시"] = pd.to_datetime(df_train["사고일시"])
df_train["year"] = df_train["사고일시"].dt.year
df_train["month"] = df_train["사고일시"].dt.month
df_train["day"] = df_train["사고일시"].dt.day
df_train["hour"] = df_train["사고일시"].dt.hour

df_test["사고일시"] = pd.to_datetime(df_test["사고일시"])
df_test["year"] = df_test["사고일시"].dt.year
df_test["month"] = df_test["사고일시"].dt.month
df_test["day"] = df_test["사고일시"].dt.day
df_test["hour"] = df_test["사고일시"].dt.hour

df_train.head()
```

```python
df_train["시군구"].unique()
```

```python
# 전부 대구이므로 구를 나눠서 변환할 필요 있음
df_train["구"] = df_train["시군구"].apply(lambda x : x.split()[1])
df_train["동"] = df_train["시군구"].apply(lambda x : x.split()[2])
df_train_drop = df_train.drop(["시군구", "ID", "사고일시"], axis=1)

df_test["구"] = df_test["시군구"].apply(lambda x : x.split()[1])
df_test["동"] = df_test["시군구"].apply(lambda x : x.split()[2])
df_test_drop = df_test.drop(["시군구", "ID", "사고일시"], axis=1)
```

```python
df_train_drop
```

## 인코딩

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df_train_drop.apply(lambda x:len(x.unique()))
```

```python
from sklearn.model_selection import train_test_split
x = df_train_drop.drop(["ECLO"], axis=1)
y = df_train_drop[["ECLO"]]
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
print(len(X_train), len(X_val))
```

```python
# 동을 빈도 인코딩 적용
df_train_drop.groupby("동")["동"].count().sort_values(ascending=False)
```

```python
mapping = df_train_drop["동"].value_counts().to_dict()
X_train["동"] = X_train["동"].map(mapping)
X_val["동"] = X_val["동"].map(mapping).fillna(0)
df_test_drop["동"] = df_test_drop["동"].map(mapping).fillna(0)
X_train["동"]
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train[["동"]] = scaler.fit_transform(X_train[["동"]])
X_val[["동"]] = scaler.transform(X_val[["동"]])
df_test_drop[["동"]] = scaler.transform(df_test_drop[["동"]])
```

```python
X_train
```

```python
df_train_drop.apply(lambda x:len(x.unique()))
```

```python
one_cols = ["요일", "기상상태", "노면상태", "사고유형", "year", "구"]
one = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
one_data = one.fit_transform(X_train[one_cols])
mul_cols = one.get_feature_names_out()
one_pd = pd.DataFrame(one_data, columns=mul_cols, index=X_train[one_cols].index)
X_train = pd.concat([X_train.drop(one_cols, axis=1), one_pd], axis=1)

one_data = one.transform(X_val[one_cols])
one_pd = pd.DataFrame(one_data, columns=mul_cols, index=X_val[one_cols].index)
X_val = pd.concat([X_val.drop(one_cols, axis=1), one_pd], axis=1)

one_data = one.transform(df_test_drop[one_cols])
one_pd = pd.DataFrame(one_data, columns=mul_cols, index=df_test_drop[one_cols].index)
df_test_drop = pd.concat([df_test_drop.drop(one_cols, axis=1), one_pd], axis=1)
```

```python
label_cols = ["도로형태"]
lbl_encoder = LabelEncoder()
X_train["도로형태"] = lbl_encoder.fit_transform(X_train["도로형태"])
```

```python
X_val["도로형태"] = lbl_encoder.transform(X_val["도로형태"])
df_test_drop["도로형태"] = lbl_encoder.transform(df_test_drop["도로형태"])
```

```python
df_test_drop.info()
```

## 학습

```python
X_val.head()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
gb = GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

gb.fit(X_train, y_train)
```

```python
y_pred = gb.predict(X_val)

print(mean_squared_error(y_pred, y_val))
print(r2_score(y_pred, y_val))
```

## 음수 값이 이상하므로 설칭

```python
gb.get_params()
```

```python
params = {
    'learning_rate': [0.1, 0.05, 0.01],
    'max_depth': [3, 4, 5],
    'min_samples_leaf': [1, 5, 10, 20],
    'min_samples_split': [2, 5, 10, 20],
    'n_estimators': [100, 200, 300, 400]
}

cv = RandomizedSearchCV(
    gb,
    param_distributions=params,
    cv=5,
    n_iter=20,
    random_state=42,
    refit=True
)
cv.fit(X_train, y_train.values.ravel())
```

```python
print(cv.best_estimator_)
print(cv.best_score_)
print(cv.best_params_)
dir(cv)
y_pred = cv.best_estimator_.predict(X_val)

print(mean_squared_error(y_pred, y_val))
print(r2_score(y_pred, y_val))
```
