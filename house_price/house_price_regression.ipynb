## 인코딩

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

y = df_drop[["SalePrice"]].values
x = df_drop.drop(["SalePrice"], axis=1)

categoric_cols = x.columns[x.dtypes == "object"].to_list()
numeric_cols = x.columns[x.dtypes != "object"].to_list()
print(len(categoric_cols) + len(numeric_cols) == len(x.columns))

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
X_train.head()
print(len(X_train), len(X_test))
print(categoric_cols)
print(numeric_cols)
```

```python
for col in categoric_cols:
  lbl_encoder = LabelEncoder()
  lbl_encoder.fit(pd.concat([X_train[col], X_test[col]]).astype(str).unique())
  X_train[col] = lbl_encoder.transform(X_train[col].astype(str))
  X_test[col] = lbl_encoder.transform(X_test[col].astype(str))

scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

X_train.head()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import RandomizedSearchCV

gb = GradientBoostingRegressor(
    n_estimators=200,
    max_depth=3,
    learning_rate=0.1,
    random_state=42
)
gb.get_params()
```

```python
parms_distri = {
    "learning_rate" : [0.01, 0.05, 0.1],
    "max_depth" : [1,2,3,4,5],
    "min_samples_leaf" : [1,2,3],
    "min_samples_split" : [2,3,4],
    "n_estimators" : [100, 200, 300]
}

rs = RandomizedSearchCV(
    gb,
    param_distributions=parms_distri,
    n_iter=100,
    cv=3,
    scoring="neg_mean_squared_error",
    refit=True
)

rs.fit(X_train, y_train)
```

```python
y_predict = rs.predict(X_test)

print("mse : ", mean_squared_error(y_test, y_predict))
print("r2 : ", r2_score(y_test, y_predict))
```
